Puisque nous avons un ensemble de données avec des caractéristiques mixtes : catégorique et numérique, nous ne pouvons pas simplement utiliser la fonction de corrélation standard (coefficient de Pearson). Nous devrons diviser les caractéristiques en tableaux séparés, et mesurer la corrélation des données numériques via Pearson, et les données catégoriques via la fonction V de Cramer.

```python
df_categorical = df.select_dtypes(include=['object'])
df_categorical = df_categorical[['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category', 'Attrition_Flag']]
df_categorical.head()
```

```python
df_numerical = df.select_dtypes(exclude=['object'])
df_numerical['Attrition_Flag'] = df['Attrition_Flag']
oh = pd.get_dummies(df_numerical['Attrition_Flag'])
df_numerical = df_numerical.drop(['Attrition_Flag'], axis=1)
df_numerical = df_numerical.drop(['CLIENTNUM'], axis=1)
df_numerical = df_numerical.join(oh)
df_numerical.head()
```

```python
from sklearn import preprocessing

label = preprocessing.LabelEncoder()
df_categorical_encoded = pd.DataFrame() 

for i in df_categorical.columns :
  df_categorical_encoded[i]=label.fit_transform(df_categorical[i])

from scipy.stats import chi2_contingency

def cramers_V(var1,var2) :
  crosstab =np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) # Cross table building
  stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
  obs = np.sum(crosstab) # Number of observations
  mini = min(crosstab.shape)-1 # Take the minimum value between the columns and the rows of the cross table
  return (stat/(obs*mini))

rows= []

for var1 in df_categorical_encoded:
  col = []
  for var2 in df_categorical_encoded :
    cramers =cramers_V(df_categorical_encoded[var1], df_categorical_encoded[var2]) # Cramer's V test
    col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
  rows.append(col)
  
cramers_results = np.array(rows)
cramerv_matrix = pd.DataFrame(cramers_results, columns = df_categorical_encoded.columns, index =df_categorical_encoded.columns)
mask = np.triu(np.ones_like(cramerv_matrix, dtype=bool))
cat_heatmap = sns.heatmap(cramerv_matrix, mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
cat_heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

Nous voyons ici que les colonnes catégoriques ne sont PAS CORRÉLÉES avec le churn du client par elles-mêmes. Cela confirme les chiffres précédents qui montrent une répartition similaire entre les clients existants et les clients désaffectés dans tous les groupes démographiques. Construire un modèle précis sans tenir compte des valeurs numériques est impossible.

Maintenant, vérifions la corrélation des valeurs numériques.

```python
from scipy import stats
num_corr=df_numerical.corr()
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(num_corr, dtype=bool))
num_heatmap = sns.heatmap(num_corr, mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
num_heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

Nous voyons maintenant de meilleures mesures de corrélation avec le taux de désabonnement du client. Dans la heatmap ci-dessus, le coefficient de corrélation des clients attrités et existants à toutes les colonnes d’entités est identique en nombre, avec des signes de miroir (+ et -). Voyons-les mieux :

```python
fig, ax=plt.subplots(ncols=2,figsize=(15, 5))

heatmap = sns.heatmap(num_corr[['Existing Customer']].sort_values(by='Existing Customer', ascending=False), ax=ax[0],vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Features Correlating with Existing Customers', fontdict={'fontsize':18}, pad=16);
heatmap = sns.heatmap(num_corr[['Attrited Customer']].sort_values(by='Attrited Customer', ascending=False), ax=ax[1],vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Features Correlating with Attrited Customers', fontdict={'fontsize':18}, pad=16);

fig.tight_layout(pad=5)
```

2.Moyenne d’ouverture à l’achat (Average Open To Buy)

4.Âge (Age)

Nous éliminons les fonctionnalités mentionnées ci-dessus pour construire un modèle précis.

```python
df_model=df
df_model=df_model.drop(['CLIENTNUM','Credit_Limit','Customer_Age','Avg_Open_To_Buy','Months_on_book','Dependent_count'],axis=1)
df_model.head()
```

Maintenant, nous convertissons les caractéristiques catégoriques en binaire avec un codage à chaud ( hot encoding)

```python
df_model['Attrition_Flag'] = df_model['Attrition_Flag'].map({'Existing Customer': 1, 'Attrited Customer': 0})
df_oh=pd.get_dummies(df_model)
df_oh['Attrition_Flag'] = df_oh['Attrition_Flag'].map({1: 'Existing Customer', 0: 'Attrited Customer'})
list(df_oh.columns)
```

# Entrainement du model

Nous formons maintenant différents classificateurs et essayons d’obtenir le modèle le plus précis. Nous utiliserons :

1. Random Forest
2. Support Vector Machine
3. Gradient Boosting

Avant de former notre modèle de prédiction, nous devrons d’abord modifier notre jeu de données déséquilibré. Nous avons vu que le client retourné ne prend que 16,1% des données.

Pour équilibrer l’ensemble de données, nous utiliserons SMOTE. SMOTE (Synthetic Minority Oversampling Technique) consiste à synthétiser des éléments pour la classe minoritaire, à partir de ceux qui existent déjà. Il fonctionne de manière aléatoire en repérant un point de la classe minoritaire et en calculant les k-voisins les plus proches pour ce point. Les points synthétiques sont ajoutés entre le point choisi et ses voisins.

D’abord nous divisons nos données de formation et de test. SMOTE sera appliqué uniquement à l’ensemble de données de formation afin que la prédiction utilise l’ensemble de données original sans points de données synthétiques.

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

X = df_oh.loc[:, df_oh.columns != 'Attrition_Flag']
y = df_oh['Attrition_Flag']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
sm = SMOTE(sampling_strategy='minority', k_neighbors=20, random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

RandomForest


```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier()
rf_clf.fit(X_train_res, y_train_res)
```

Support Vector Machine

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
svm_clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
svm_clf.fit(X_train_res, y_train_res)
```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_clf=GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=42)
gb_clf.fit(X_train_res, y_train_res)
```

# Prediction des models

Première prediction : 

```python
y_rf=rf_clf.predict(X_test)
y_svm=svm_clf.predict(X_test)
y_gb=gb_clf.predict(X_test)
```

Première evaluation : 

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

fig,ax=plt.subplots(ncols=3, figsize=(20,6))

ConfusionMatrixDisplay.from_estimator(rf_clf, X_test, y_test, ax=ax[0]).plot()
ax[0].title.set_text('Random Forest')

ConfusionMatrixDisplay.from_estimator(svm_clf, X_test, y_test, ax=ax[1]).plot()
ax[1].title.set_text('Support Vector Machine')

ConfusionMatrixDisplay.from_estimator(gb_clf, X_test, y_test, ax=ax[2]).plot()
ax[2].title.set_text('Gradient Boosting')

fig.tight_layout(pad=5)
```

```python
from sklearn.metrics import classification_report, recall_score, precision_score, f1_score
print('Random Forest Classifier')
print(classification_report(y_test, y_rf))
print('------------------------')
print('Support Vector Machine')
print(classification_report(y_test, y_svm))
print('------------------------')
print('Gradient Boosting')
print(classification_report(y_test, y_gb))
```

# Tuning hyperparamètre

```python
from sklearn.model_selection import RandomizedSearchCV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf_clf_opt= RandomForestClassifier(n_estimators=750, min_samples_split=2, min_samples_leaf=1, 
                            max_features='log2', max_depth=50, bootstrap=False)
rf_clf_opt.fit(X_train_res,y_train_res)
y_rf_opt=rf_clf_opt.predict(X_test)
print('Random Forest Classifier (Optimized)')
print(classification_report(y_test, y_rf_opt))
```

```python
from sklearn.metrics import confusion_matrix

# Calculer la matrice de confusion
conf_matrix = confusion_matrix(y_test, y_rf_opt)

# Afficher la matrice de confusion
_rf_opt = ConfusionMatrixDisplay(conf_matrix)
_rf_opt.plot()
plt.title('Confusion Matrix - Random Forest Classifier (Optimized)')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

param_test1 = {'n_estimators':range(20,81,10)}
gsearch1 = GridSearchCV(
estimator = GradientBoostingClassifier(learning_rate=1.0, min_samples_split=500,min_samples_leaf=50,max_depth=8,max_features='sqrt',subsample=0.8,random_state=10), 
param_grid = param_test1, scoring='roc_auc',n_jobs=4, cv=5)
gsearch1.fit(X_train_res,y_train_res)
print(gsearch1.best_params_)
```

```python
gb_clf_opt=GradientBoostingClassifier(n_estimators=80,learning_rate=1.0, min_samples_split=500,min_samples_leaf=50,max_depth=8,max_features='sqrt',subsample=0.8,random_state=10)
gb_clf_opt.fit(X_train_res,y_train_res)
y_gb_opt=gb_clf_opt.predict(X_test)
print('Gradient Boosting (Optimized)')
print(classification_report(y_test, y_gb_opt))
```

```python
from sklearn.metrics import confusion_matrix

# Calculer la matrice de confusion
conf_matrix_gb = confusion_matrix(y_test, y_gb_opt)

# Afficher la matrice de confusion
_gb_opt = ConfusionMatrixDisplay(conf_matrix_gb)
_gb_opt.plot()
plt.title('Confusion Matrix - Gradient Boosting (Optimized)')
plt.show()
```
