Puisque nous avons un ensemble de donnÃ©es avec des caractÃ©ristiques mixtesÂ : catÃ©gorique et numÃ©rique, nous ne pouvons pas simplement utiliser la fonction de corrÃ©lation standard (coefficient de Pearson). Nous devrons diviser les caractÃ©ristiques en tableaux sÃ©parÃ©s, et mesurer la corrÃ©lation des donnÃ©es numÃ©riques via Pearson, et les donnÃ©es catÃ©goriques via la fonction V de Cramer.

```python
df_categorical = df.select_dtypes(include=['object'])
df_categorical = df_categorical[['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category', 'Attrition_Flag']]
df_categorical.head()
```

```python
df_numerical = df.select_dtypes(exclude=['object'])
df_numerical['Attrition_Flag'] = df['Attrition_Flag']
oh = pd.get_dummies(df_numerical['Attrition_Flag'])
df_numerical = df_numerical.drop(['Attrition_Flag'], axis=1)
df_numerical = df_numerical.drop(['CLIENTNUM'], axis=1)
df_numerical = df_numerical.join(oh)
df_numerical.head()
```

```python
from sklearn import preprocessing

label = preprocessing.LabelEncoder()
df_categorical_encoded = pd.DataFrame() 

for i in df_categorical.columns :
  df_categorical_encoded[i]=label.fit_transform(df_categorical[i])

from scipy.stats import chi2_contingency

def cramers_V(var1,var2) :
  crosstab =np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) # Cross table building
  stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
  obs = np.sum(crosstab) # Number of observations
  mini = min(crosstab.shape)-1 # Take the minimum value between the columns and the rows of the cross table
  return (stat/(obs*mini))

rows= []

for var1 in df_categorical_encoded:
  col = []
  for var2 in df_categorical_encoded :
    cramers =cramers_V(df_categorical_encoded[var1], df_categorical_encoded[var2]) # Cramer's V test
    col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
  rows.append(col)
  
cramers_results = np.array(rows)
cramerv_matrix = pd.DataFrame(cramers_results, columns = df_categorical_encoded.columns, index =df_categorical_encoded.columns)
mask = np.triu(np.ones_like(cramerv_matrix, dtype=bool))
cat_heatmap = sns.heatmap(cramerv_matrix, mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
cat_heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

Nous voyons ici que les colonnes catÃ©goriques ne sont PAS CORRÃ‰LÃ‰ES avec le churn du client par elles-mÃªmes. Cela confirme les chiffres prÃ©cÃ©dents qui montrent une rÃ©partition similaire entre les clients existants et les clients dÃ©saffectÃ©s dans tous les groupes dÃ©mographiques. Construire un modÃ¨le prÃ©cis sans tenir compte des valeurs numÃ©riques est impossible.

Maintenant, vÃ©rifions la corrÃ©lation des valeurs numÃ©riques.

```python
from scipy import stats
num_corr=df_numerical.corr()
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(num_corr, dtype=bool))
num_heatmap = sns.heatmap(num_corr, mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
num_heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

Nous Ã©liminons les fonctionnalitÃ©s mentionnÃ©es ci-dessus pour construire un modÃ¨le prÃ©cis.

```python
df_model=df
df_model=df_model.drop(['CLIENTNUM','Credit_Limit','Customer_Age','Avg_Open_To_Buy','Months_on_book','Dependent_count'],axis=1)
df_model.head()
```

Maintenant, nous convertissons les caractÃ©ristiques catÃ©goriques en binaire avec un codage Ã  chaud ( hot encoding)

```python
df_model['Attrition_Flag'] = df_model['Attrition_Flag'].map({'Existing Customer': 1, 'Attrited Customer': 0})
df_oh=pd.get_dummies(df_model)
df_oh['Attrition_Flag'] = df_oh['Attrition_Flag'].map({1: 'Existing Customer', 0: 'Attrited Customer'})
list(df_oh.columns)
```

# Entrainement du model

Nous formons maintenant diffÃ©rents classificateurs et essayons dâ€™obtenir le modÃ¨le le plus prÃ©cis. Nous utiliseronsÂ :

1. Random Forest
2. Support Vector Machine
3. Gradient Boosting

Avant de former notre modÃ¨le de prÃ©diction, nous devrons dâ€™abord modifier notre jeu de donnÃ©es dÃ©sÃ©quilibrÃ©. Nous avons vu que le client retournÃ© ne prend que 16,1% des donnÃ©es.

Pour Ã©quilibrer lâ€™ensemble de donnÃ©es, nous utiliserons SMOTE. SMOTE (Synthetic Minority Oversampling Technique) consiste Ã  synthÃ©tiser des Ã©lÃ©ments pour la classe minoritaire, Ã  partir de ceux qui existent dÃ©jÃ . Il fonctionne de maniÃ¨re alÃ©atoire en repÃ©rant un point de la classe minoritaire et en calculant les k-voisins les plus proches pour ce point. Les points synthÃ©tiques sont ajoutÃ©s entre le point choisi et ses voisins.

Dâ€™abord nous divisons nos donnÃ©es de formation et de test. SMOTE sera appliquÃ© uniquement Ã  lâ€™ensemble de donnÃ©es de formation afin que la prÃ©diction utilise lâ€™ensemble de donnÃ©es original sans points de donnÃ©es synthÃ©tiques.

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

X = df_oh.loc[:, df_oh.columns != 'Attrition_Flag']
y = df_oh['Attrition_Flag']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
sm = SMOTE(sampling_strategy='minority', k_neighbors=20, random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

RandomForest


```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier()
rf_clf.fit(X_train_res, y_train_res)
```

Support Vector Machine

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
svm_clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
svm_clf.fit(X_train_res, y_train_res)
```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_clf=GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=42)
gb_clf.fit(X_train_res, y_train_res)
```

# Prediction des models

PremiÃ¨re prediction : 

```python
y_rf=rf_clf.predict(X_test)
y_svm=svm_clf.predict(X_test)
y_gb=gb_clf.predict(X_test)
```

PremiÃ¨re evaluation : 

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

fig,ax=plt.subplots(ncols=3, figsize=(20,6))

ConfusionMatrixDisplay.from_estimator(rf_clf, X_test, y_test, ax=ax[0]).plot()
ax[0].title.set_text('Random Forest')

ConfusionMatrixDisplay.from_estimator(svm_clf, X_test, y_test, ax=ax[1]).plot()
ax[1].title.set_text('Support Vector Machine')

ConfusionMatrixDisplay.from_estimator(gb_clf, X_test, y_test, ax=ax[2]).plot()
ax[2].title.set_text('Gradient Boosting')

fig.tight_layout(pad=5)
```

```python
from sklearn.metrics import classification_report, recall_score, precision_score, f1_score
print('Random Forest Classifier')
print(classification_report(y_test, y_rf))
print('------------------------')
print('Support Vector Machine')
print(classification_report(y_test, y_svm))
print('------------------------')
print('Gradient Boosting')
print(classification_report(y_test, y_gb))
```

# Tuning hyperparamÃ¨tre

```python
from sklearn.model_selection import RandomizedSearchCV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf_clf_opt= RandomForestClassifier(n_estimators=750, min_samples_split=2, min_samples_leaf=1, 
                            max_features='log2', max_depth=50, bootstrap=False)
rf_clf_opt.fit(X_train_res,y_train_res)
y_rf_opt=rf_clf_opt.predict(X_test)
print('Random Forest Classifier (Optimized)')
print(classification_report(y_test, y_rf_opt))
```

```python
from sklearn.metrics import confusion_matrix

# Calculer la matrice de confusion
conf_matrix = confusion_matrix(y_test, y_rf_opt)

# Afficher la matrice de confusion
_rf_opt = ConfusionMatrixDisplay(conf_matrix)
_rf_opt.plot()
plt.title('Confusion Matrix - Random Forest Classifier (Optimized)')
plt.show()
```
